**Build a pyramid**

```python
rows = 4
for x in range (rows):
    for y in range (x):
        print ('#', end = ' ')
    print(' ')
```

```python
rows = 4
for x in range (1, rows+1):
    for y in range (x):
        print ('#', end = ' ')
    print(' ')
```
